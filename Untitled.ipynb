```python
def normalise_data(data):
    normalized_df = pd.DataFrame()
    for column in data.columns:
        if 'Quat_W' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'Quat_X' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'Quat_Y' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'Quat_Z' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'FreeAcc_X' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'FreeAcc_Y' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
        if 'FreeAcc_Z' in column:
            normalized_column = data[column] / data[column].max()  # Normalize the column
            normalized_df[column] = normalized_column  # Add the normalized column to the new DataFrame
    
    return normalized_df
        
            
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
X = merged_df[['Quat_W DOT1_','Quat_X DOT1_','Quat_Y DOT1_','Quat_Z DOT1_','FreeAcc_X DOT1_','FreeAcc_Y DOT1_','FreeAcc_Z DOT1_']]
```

```python
y_DOT2 = merged_df[['Quat_W DOT 2',
    'Quat_X DOT 2',
    'Quat_Y DOT 2',
    'Quat_Z DOT 2',
    'FreeAcc_X DOT 2',
    'FreeAcc_Y DOT 2',
    'FreeAcc_Z DOT 2']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y_DOT2,test_size=0.2,random_state=42)
```

```python
X_train = np.array(X_train).reshape(X_train.shape[0], 1, X_train.shape[1])
```

```python
model = Sequential()
model.add(LSTM(50,input_shape=(1,7)))
model.add(Dense(7))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train,y_train,epochs=50,batch_size=32,verbose=1)
```

```python
mse = model.evaluate(X_test, y_test)
print("Mean Squared Error:", mse)
```
