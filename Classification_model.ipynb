```python
def merge_data(path):
    merge_df = pd.DataFrame()
    for files in os.listdir(path):
        if '.csv' in files:
            print(files)
            name_to_add = files[5:11]
            print(name_to_add)
            print(path+files)
            df = pd.read_csv(path+files, skiprows=10)
            df.columns = [col + name_to_add for col in df.columns]
            print(df)
            merge_df = pd.concat([merge_df, df], axis=1)
            print(merge_df)
    return merge_df
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = merged_df[['Quat_W DOT 2', 'Quat_X DOT 2', 'Quat_Y DOT 2', 'Quat_Z DOT 2',
       'FreeAcc_X DOT 2', 'FreeAcc_Y DOT 2', 'FreeAcc_Z DOT 2', 'Quat_W DOT 3',
       'Quat_X DOT 3', 'Quat_Y DOT 3', 'Quat_Z DOT 3', 'FreeAcc_X DOT 3',
       'FreeAcc_Y DOT 3', 'FreeAcc_Z DOT 3', 'Quat_W DOT 4', 'Quat_X DOT 4',
       'Quat_Y DOT 4', 'Quat_Z DOT 4', 'FreeAcc_X DOT 4', 'FreeAcc_Y DOT 4',
       'FreeAcc_Z DOT 4', 'Quat_W DOT 5', 'Quat_X DOT 5', 'Quat_Y DOT 5',
       'Quat_Z DOT 5', 'FreeAcc_X DOT 5', 'FreeAcc_Y DOT 5', 'FreeAcc_Z DOT 5',
       'Quat_W DOT1_', 'Quat_X DOT1_', 'Quat_Y DOT1_', 'Quat_Z DOT1_',
       'FreeAcc_X DOT1_', 'FreeAcc_Y DOT1_', 'FreeAcc_Z DOT1_']], merged_df['Status']
```

```python
X_train, X_test , y_train,y_test = train_test_split(X,y, test_size=.2, random_state=42)
```

```python
model = SGDClassifier(loss='log_loss',alpha=0.01,learning_rate='optimal',random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accracy = accuracy_score(y_test,y_pred)
```

```python
print("Initial Model Accuracy:", accracy)
```

```python
X_new, y_new = X_test[:5],y_test[:5]
```

```python
model.partial_fit(X_new,y_new)
```

```python
y_pred_updated = model.predict(X_test)
```

```python
accuracy_updated = accuracy_score(y_test, y_pred_updated)
print("Updated Model Accuracy:", accuracy_updated)
```
